# Bitsandbytes quantization

```python
"""Mini-CPM
"""

from transformers import AutoModel, AutoTokenizer
from transformers import BitsAndBytesConfig

model_path = "../weights/hf/MiniCPM-Llama3-V-2_5"

q_config = BitsAndBytesConfig(
             load_in_8bit=True,
             llm_int8_skip_modules=['out_proj', 'kv_proj', 'lm_head'])
llm = AutoModel.from_pretrained(model_path, trust_remote_code=True, quantization_config=q_config)
tokenizer = AutoTokenizer.from_pretrained(model_path, trust_remote_code=True)

# save quantized model
llm.save_pretrained(model_path + "_int8")
tokenizer.save_pretrained(model_path + "_int8")
```
